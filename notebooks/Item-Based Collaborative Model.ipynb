```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import pairwise_distances 

data_path = '../data/21B_tag_views_dataset.csv'
```

# Item-based Collaborative Filtering

## Data preparation

Data loading and user/tag freq matrix

```python
data = pd.read_csv(data_path)
```

```python
tag_count_df = data.groupby(['user_id', 'tag_id']).agg({'tag_id': 'count'}).rename(columns={'tag_id':'tag_count'})
tag_count_df.head()
```

```python
tag_count_df = tag_count_df.reset_index()
tags_user_df = tag_count_df.pivot(index='tag_id', columns='user_id', values='tag_count').fillna(0)
tags_user_df.head()
```

### Model engine with cosine distance
Example model of item similarty based on cosine distance

```python
item_distances = pairwise_distances(tags_user_df.values, metric='cosine')
item_distances.shape
```

```python
item_distances
```

Convert the distances to "similarity" scores

```python
item_similarity = 1 - item_distances
```

```python
item_similarity
```

Set the same tag similarities to zero:

```python
np.fill_diagonal(item_similarity, 0)
```

```python
item_similarity
```

```python
item_similarity = pd.DataFrame(item_similarity, columns=tags_user_df.index, index=tags_user_df.index)
item_similarity.head()
```

### Tag values for "ff0d3fb21c00bc33f71187a2beec389e9eff5332"

```python
tag1 = "ff0d3fb21c00bc33f71187a2beec389e9eff5332"
```

```python
item_similarity.loc[tag1].sort_values(ascending=False)[:10]
```

```python
def get_tag_predictions(item_similarity_scores, tag,top_n):
    items_keep = item_similarity_scores.loc[tag][item_similarity_scores.loc[tag]>0]
    return items_keep.sort_values(ascending=False)[:top_n]
```

```python
predicted_tags = get_tag_predictions(item_similarity, tag1, 10)
predicted_tags
```

### Tag popularity model

Another approach is to simply get the most viewed tag from the users that viewed our target tag. Here we will construct a simple model to retreive those tags.

```python
tags_user_df.head()
```

We get invert this matrix for to take advantage of the user index:

```python
user_tags_df = tags_user_df.T.copy()
```

With the following function we can retreive the most popular tags viewed by the users that viewed our tag

```python
def get_popular_tags(tag, top_n=10):
    predictions = None
    # Filter to get the users which viewed our tag
    tag_viewers = tags_user_df.loc[tag][tags_user_df.loc[tag] > 0]
    
    # Loop to retreive the tags viewed by each user
    for user in tag_viewers.index:
        # Get the tags viewed by the current loop user
        tags_viewed = user_tags_df.loc[user][user_tags_df.loc[user]>0]
        if predictions is None: # In the first case we initialize the predictions variable
            predictions = tags_viewed
            continue
        # Sum the tag values
        predictions = predictions.add(tags_viewed, fill_value=0)
    # Filtering zero elements, just in case 
    predictions = predictions[predictions>0]
    return predictions.sort_values(ascending=False)[:top_n]
```

```python
predictions = get_popular_tags('00410345e6d60633a211ebd3755d5c89ea7b5297')
```
